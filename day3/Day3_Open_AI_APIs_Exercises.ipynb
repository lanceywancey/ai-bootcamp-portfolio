### Task 3.3 — The Full `MemoryBot` Class

**Your Task:** No blanks here — run the cell as-is and test it. Notice that `tutor` and `coach` have **completely separate** and independent histories.

```python
# Task 3.3 — MemoryBot class (no blanks — run and observe)

class MemoryBot:
    def __init__(self, personality='helpful'):
        prompts = {
            'helpful': 'You are a friendly, helpful assistant.',
            'tutor':   'You are a patient CS tutor. Use simple examples. Always check understanding.',
            'coach':   'You are an energetic life coach. Always motivate. Start every reply with an emoji.',
        }
        self.system    = prompts.get(personality, prompts['helpful'])
        self.history   = []
        self.turn      = 0
        self.total_tok = 0

    def ask(self, question):
        self.turn += 1
        self.history.append({'role': 'user', 'content': question})
        messages = [{'role': 'system', 'content': self.system}] + self.history
        response = client.chat.completions.create(
            model      = 'gpt-4o-mini',
            messages   = messages,
            max_tokens = 200
        )
        answer = response.choices[0].message.content
        self.total_tok += response.usage.total_tokens
        self.history.append({'role': 'assistant', 'content': answer})
        return answer

    def status(self):
        cost = self.total_tok * 0.0000002
        return f'Turn {self.turn} | {len(self.history)} messages | {self.total_tok} tokens | ${cost:.6f}'


# Test tutor personality
print('=== TUTOR BOT ===')
tutor = MemoryBot(personality='tutor')
print(tutor.ask('Explain what a variable is.'))
print(tutor.status())
print(tutor.ask('Can you give me a quiz question about variables?'))
print(tutor.status())
print()

# Test coach personality
print('=== COACH BOT ===')
coach = MemoryBot(personality='coach')
print(coach.ask('I feel like giving up on programming. It is too hard.'))
print(coach.status())
```
